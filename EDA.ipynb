```python
# Define a dictionary to map the changes
bu_mapping = {
    'Major Projects': 'Major Projects',
    'Infrastructure': 'Infrastructure',
    'Rail': 'Rail',
    'Building': 'Building',
    'Finance & Group Services': 'Corporate',
    'TEK': 'TEK',
    'HSE': 'Corporate',
    'WA Region': 'Corporate',
    'People': 'Corporate',
    'Strategy': 'Corporate',
    'ICT': 'Corporate',
    'D&I': 'D&I',
    'Customer & Corp Affairs': 'Corporate',
    'Commercial': 'Corporate',
    'Chief Operations Office': 'Corporate',
    'Executive Management': 'Corporate'
}

# Create a new column 'BU1' based on the mapping
df['BU1'] = df['BU'].replace(bu_mapping)
```

```python
# Filter out active employees
active_employees_df = df[df["Employee Status"] == "Active"]
active_employees_df
```

```python
# Check for duplicate Person IDs
duplicate_person_ids = active_employees_df[active_employees_df.duplicated(subset='Person ID', keep=False)]

# If there are duplicates, count the occurrences
if not duplicate_person_ids.empty:
    duplicate_counts = duplicate_person_ids['Person ID'].value_counts()
    print("Duplicate Person IDs with their occurrence count:")
    print(duplicate_counts)
else:
    print("No duplicate Person IDs found.")
```

```python
# Sort the DataFrame by reporting date in descending order
active_employees_df_sorted = active_employees_df.sort_values(by='Reporting Date', ascending=False)

# Drop duplicate entries keeping the first occurrence (latest reporting date)
active_employees_df = active_employees_df_sorted.drop_duplicates(subset='Person ID', keep='first')

# Print the DataFrame with unique entries
active_employees_df
```

```python
import matplotlib.pyplot as plt

# Filter active employees who are managers
manager_df = active_employees_df[active_employees_df["Manager/No Manager"] == "Manager"]

# Group by 'BU' and count the number of managers in each business unit
manager_count = manager_df.groupby('BU1').size()

# Plot histogram
ax = manager_count.plot(kind='bar', figsize=(8, 6), color='red')
plt.title('Number of Managers in Each Business Unit')
plt.xlabel('Business Unit')
plt.ylabel('Number of Managers')
plt.xticks(rotation=45, ha='right')
# Add numbers above each bar
for i, count in enumerate(manager_count):
    ax.text(i, count + 0.1, str(count), ha='center')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Your data (employees_by_bu_gender_employment) here

# Plot donut charts for each business unit
for bu, data in employees_by_bu_gender_employment.groupby(level=0):
    fig, axes = plt.subplots(nrows=1, ncols=len(data), figsize=(15, 6), subplot_kw=dict(aspect="equal"))
    fig.suptitle(f'Number of Employees by Employment Type and Gender in {bu}', fontsize=16)

    for ax, (_, row) in zip(axes, data.iterrows()):
        sizes = row.values
        labels = row.index
        colors = ['#1f77b4', 'red', 'grey', '#d62728', '#9467bd']
        explode = (0.1, 0, 0, 0, 0)  # explode the 1st slice

        ax.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140, explode=explode)
        ax.set_title(row.name, fontsize=12)

        # Draw a circle at the center of pie to make it a donut
        centre_circle = plt.Circle((0,0),0.70,fc='white')
        ax.add_patch(centre_circle)

    plt.tight_layout()
    plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Filter employees in critical roles
critical_employees = df[df['Critical / Core'] == 'Critical']

# Define the order of age groups and generations
age_groups_order = ['Less than 20 yrs', '20-30 yrs', '30-40 yrs', '40-50 yrs', '50 and more']
generations_order = ['Baby Bommer', 'Millenial', 'Gen X', 'Gen Z']

# Set the style of seaborn
sns.set_style("whitegrid")

# Create the cluster plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=critical_employees, x='Age (Group)', y='Generation', hue='Person ID',
                hue_order=age_groups_order, style='Generation', style_order=generations_order,
                palette='coolwarm', markers=['o', 's', 'p', 'D'], legend='full')
plt.title('Critical Employees by Age Group and Generation')
plt.xlabel('Age Group')
plt.ylabel('Generation')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```
