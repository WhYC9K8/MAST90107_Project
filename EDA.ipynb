```python
import matplotlib.pyplot as plt

# Filter active employees who are managers
manager_df = active_employees_df[active_employees_df["Manager/No Manager"] == "Manager"]

# Group by 'BU' and count the number of managers in each business unit
manager_count = manager_df.groupby('BU1').size()

# Plot histogram
ax = manager_count.plot(kind='bar', figsize=(8, 6), color='red')
plt.title('Number of Managers in Each Business Unit')
plt.xlabel('Business Unit')
plt.ylabel('Number of Managers')
plt.xticks(rotation=45, ha='right')
# Add numbers above each bar
for i, count in enumerate(manager_count):
    ax.text(i, count + 0.1, str(count), ha='center')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Plot donut charts for each business unit
for bu, data in employees_by_bu_gender_employment.groupby(level=0):
    fig, axes = plt.subplots(nrows=1, ncols=len(data), figsize=(15, 6), subplot_kw=dict(aspect="equal"))
    fig.suptitle(f'Number of Employees by Employment Type and Gender in {bu}', fontsize=16)

    for ax, (_, row) in zip(axes, data.iterrows()):
        sizes = row.values
        labels = [row.name.split()[1]] * len(row)  # Extract employment type and replicate for each row
        colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd']
        explode = (0.1, 0, 0, 0, 0)  # explode the 1st slice

        ax.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140, explode=explode)
        ax.set_title(row.name, fontsize=12)

        # Draw a circle at the center of pie to make it a donut
        centre_circle = plt.Circle((0,0),0.70,fc='white')
        ax.add_patch(centre_circle)

    plt.tight_layout()
    plt.show()
```
